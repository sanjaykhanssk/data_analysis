```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("data/diamonds.csv")
```

```python
df.head()
```

```python
plt.figure()
g=sns.scatterplot(x="carat" , y="price" , data=df)
g.set(ylim=(0 , df.price.quantile(0.80)))
g.set(xlim=(0 , df.carat.quantile(0.80)))

plt.show()
```

```python
plt.figure()
g=sns.scatterplot(x="carat" , y="price" , data=df)
sns.regplot(x="carat" , y="price" , data=df , color="orange" ,scatter=False)
g.set(ylim=(0 , df.price.quantile(0.99)))
g.set(xlim=(0 , df.carat.quantile(0.99)))

plt.show()
```
