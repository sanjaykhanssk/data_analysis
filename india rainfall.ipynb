# Filling Na values

## Filling NaN for months

```python
def fillna_months(li , df):
    for i in li:
        df[i].fillna(df[i].mean() , inplace=True)
    return df
li = ["JAN" , "FEB" , "MAR" , "APR" , "MAY" , "JUN" , "JUL" , "AUG" , "SEP" , "OCT" , "NOV" , "DEC" ]
df2 = fillna_months(li , df)
```

```python
df2.isna().sum()
```

```python
df2.head()
```

```python
df = df2.copy()
```

```python
del df2
```

## Filling in cuts 

```python
df["Jan-Feb"].fillna(df["JAN"] + df["FEB"] , inplace=True)
df["Mar-May"].fillna(df["MAR"] + df["APR"]+df["MAY"] , inplace=True)
```

```python
df["Jun-Sep"].fillna(df["JUN"] + df["JUL"]+df["AUG"]+df["SEP"] , inplace=True)
df["Oct-Dec"].fillna(df["OCT"] + df["NOV"]+df["DEC"] , inplace=True)
```

## filling NaN for Annual

```python
df["ANNUAL"].fillna(df["Jan-Feb"]+df["Mar-May"]+df["Jun-Sep"]+df["Oct-Dec"] , inplace=True)
```

### Average Rainfall all over india in 1990 - 2015

```python
grp_rf = df3.groupby("SUBDIVISION")[["YEAR" , "ANNUAL"]]
```

```python
grp_rf = grp_rf.mean()["ANNUAL"]
```

```python
grp_rf = grp_rf.sort_values(ascending=False)
```

```python
plt.figure(figsize=(20,20))
plt.xticks(rotation=90)
plt.title("Average rainfall from 1990 - 2015 all over india")
sns.barplot(y=grp_rf.index , x=grp_rf.values , palette="winter" , orient="h")
plt.show()
```

### A best place to spend time in summer 

```python
grp_rf = df3.groupby("SUBDIVISION")["Mar-May"]
```

```python
grp_rf = grp_rf.mean().sort_values(ascending=False)
```

```python
grp_rf
```

```python
plt.figure(figsize=(20,15))
plt.xticks(rotation=90)
plt.title("Best Place to spend Summer in India")
sns.barplot(y=grp_rf.index , x=grp_rf.values , palette="summer" , orient="h")
plt.show()
```
